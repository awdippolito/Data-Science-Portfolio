```python
#Import libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# Read in data

#data = pd.read_csv("/kaggle/input/global-cost-of-living/cost-of-living_v2.csv")
data = pd.read_csv("cost-of-living_v2.csv")

print("Data has \033[1m{} rows\033[0m and \033[1m{} columns\033[0m \n".format(data.shape[0], data.shape[1]))


# Determine number of missing values per column

print("Data has the following number of \033[1mmissing values\033[0m from each column.\n\033[1m{}\033[0m".format(data.isnull().sum()))


# Look at data summary and the head rows
print("\n\033[1mOverall data summary below\033[0m")
display(data.describe(include='all'))
print("\n\033[1mFirst 5 rows of data below\033[0m")
display(data.head())
```

## Data Processing

We can see that every numerical value has some missing values and there might be some duplicate city entries (Victoria has a frequency of 4 in the city column).  This could be different cities with the same name instead of duplicates.  Let's look into Victoria first and see what we find.

```python
data[data['city']=='Victoria']
```

Looks like it is different cities with the same name in different countries and not duplicates but we should do a more rigourous check just in case.  We can also see a lot of NaN values for Victoria, Chile and Victoria, Argentina.

```python
# Create a new column combining city and country.  I'll also strip white space before and after just in case.
data['city_country'] = data['city'].str.strip() + ', ' + data['country'].str.strip()

# Next I'll check the number of unique pairings.  If this number matches the number of rows in the dataset then there are
# no duplicates.
data.pop('city_country').nunique()
```

There are 4956 different city, country pairs and 4956 rows in the dataset.  This confirms that we do not have any duplicate cities.

Next I want to do something about all the NaN values.  I'll create an additional column that counts how many NaN values we have in each row.

```python
data['NaN_count'] = data.iloc[:,2:57].isnull().sum(axis=1)

data.head()
```

Next let's look at a histogram of the NaN count for all cities.  I will also utilize the data_quality column to see it's relationship with NaN count. 

```python
axes = sns.histplot(x=data['NaN_count'], hue=data['data_quality'], multiple='stack', binwidth=4)

axes.set(xlabel='NaN count',
       ylabel='Count',
       title='Histogram of Cities NaN Count')

plt.show()
```

The figure above shows the following:
- The majority of the data has a data_quality of 0
- Data entries with a quality of 1 never have more than 3 missing values
- Data entries with a quality of 0 have a much larger range of missing value count

Based on the above, the next question is, how should we handle these cities with high NaN values and data_quality=0?  There's always a balance between cutting useful data and keeping bad data that provides inaccurate results.  In this case I'm going to drop all rows that have more than 3 missing values.  This still gives me a large number of cities to work with (~2200) and I am not cutting any city that Numbeo has given a data quality of 1.  I am still keeping a good chunk of data_quality=0 cities but since they have almost every field filled out I believe this data likely still contains some valuable information.

Below is a couple of donut plots to compare the % of cities belonging to a certain country before and after the row drop from the data.

```python
# Drop rows if they have more than 3 missing values
qual_data = data.drop(data[data['NaN_count'] > 3].index)

# Group raw data and data after row drop by city count per country
country_data = pd.DataFrame(data.groupby('country')['city'].count().sort_values(ascending=False))
country_qdata = pd.DataFrame(qual_data.groupby('country')['city'].count().sort_values(ascending=False))

# Plan to create a pair of donut charts to compare the representation of the main countries before and after the drop.
# Charts will only look at the top 12 countries after the drop
countries = list(country_qdata.iloc[0:12].index)

# Label any country outside of the top 12 as 'Other'
country_data['chart_country'] = country_data.index.map(lambda x: x if x in countries else 'Other')
country_qdata['chart_country'] = country_qdata.index.map(lambda x: x if x in countries else 'Other')

countries.append('Other')

# Group datasets again by country, this time summing the 'Other' labels
country_data = pd.DataFrame(country_data.groupby('chart_country')['city'].sum()).reindex(countries)
country_qdata = pd.DataFrame(country_qdata.groupby('chart_country')['city'].sum()).reindex(countries)

fig, axes = plt.subplots(1, 2, figsize=(18,18))

# Colors from colorbrewer2.org
colors = ['#a6cee3','#1f78b4', '#b2df8a', '#33a02c', '#fb9a99', '#e31a1c', '#fdbf6f', '#ff7f00', '#cab2d6', '#6a3d9a', '#ffff99', '#b15928', 'grey']

axes[0].pie(country_data['city'], labels = countries, colors=colors, autopct='%1.1f%%', pctdistance=0.85, wedgeprops=dict(width=0.3), textprops={'weight': 'bold'})
axes[0].set_title('% of Cities by Country in Raw Data', fontweight='bold', fontsize='x-large', y=1.05)
plt.text(-3.35,0.0, 'Total cities = %0.0f'%data.shape[0], fontsize=12, fontweight='bold')
plt.text(-3.41,-0.1, 'Total countries = %0.0f'%data['country'].nunique(), fontsize=12, fontweight='bold')

axes[1].pie(country_qdata['city'], labels = countries, colors=colors, autopct='%1.1f%%', pctdistance=0.85, wedgeprops=dict(width=0.3), textprops={'weight': 'bold'})
axes[1].set_title('% of Cities by Country in Data after Row Drop', fontweight='bold', fontsize='x-large', y=1.05)
plt.text(-0.34,0.0, 'Total cities = %0.0f'%qual_data.shape[0], fontsize=12, fontweight='bold')
plt.text(-0.4,-0.1, 'Total countries = %0.0f'%qual_data['country'].nunique(), fontsize=12, fontweight='bold')

plt.show()
```

This has cut a little over half (~54%) of the rows from our dataset and altered the proportion of cities by country and dropped the total number of countries represented from 215 to 177.  The country with the biggest change is the United States going from 21.7% to 13.9%.

Overall this appears to be a net positive for the global representation of the dataset.  There is still a large bias in data from the United States but it has been decreased after the row drop.  There is still a large underrepresentation in countries with large populations such as India and China.

One downside to this row drop is that we may have lost data for smaller cities and be primarily left with large city centers.  Further in depth analysis would be required to confirm this but that is outside of the current scope of this EDA.

Therefore, we must keep these overall biases in mind when looking at the global data distributions plotted below.

```python
fig, axes = plt.subplots(2, 3, figsize=(18,10))

fig.suptitle('Global Distribution of 6 Price Categories', fontweight='bold', fontsize='xx-large')

sns.violinplot(ax=axes[0,0], data=qual_data[qual_data['x2'].notnull()], y='x2', cut=0)
axes[0,0].set_ylabel('$USD', fontweight='bold')
axes[0,0].set_title('Mid-range Restaurant Meal for 2', fontweight='bold')
axes[0,0].set_yticklabels(axes[0,0].get_yticks(), fontweight='bold')

sns.violinplot(ax=axes[0,1], data=qual_data[qual_data['x24'].notnull()], y='x24', cut=0)
axes[0,1].set_ylabel('$USD', fontweight='bold')
axes[0,1].set_title('Bottle of Mid-range Wine', fontweight='bold')
axes[0,1].set_yticklabels(axes[0,1].get_yticks(), fontweight='bold')

sns.violinplot(ax=axes[0,2], data=qual_data[qual_data['x10'].notnull()], y='x10', cut=0)
axes[0,2].set_ylabel('$USD', fontweight='bold')
axes[0,2].set_title('Loaf of Fresh White Bread', fontweight='bold')
axes[0,2].set_yticklabels(axes[0,2].get_yticks(), fontweight='bold')

sns.violinplot(ax=axes[1,0], data=qual_data[qual_data['x33'].notnull()], y='x33', cut=0)
axes[1,0].set_ylabel('$USD', fontweight='bold')
axes[1,0].set_title('1 litre of Gasoline', fontweight='bold')
axes[1,0].set_yticklabels(axes[1,0].get_yticks(), fontweight='bold')

sns.violinplot(ax=axes[1,1], data=qual_data[qual_data['x38'].notnull()], y='x38', cut=0)
axes[1,1].set_ylabel('$USD', fontweight='bold')
axes[1,1].set_title('60 Mbps or More Internet', fontweight='bold')
axes[1,1].set_yticklabels(axes[1,1].get_yticks(), fontweight='bold')

sns.violinplot(ax=axes[1,2], data=qual_data[qual_data['x41'].notnull()], y='x41', cut=0)
axes[1,2].set_ylabel('$USD', fontweight='bold')
axes[1,2].set_title('Cinema International Release', fontweight='bold')
axes[1,2].set_yticklabels(axes[1,2].get_yticks(), fontweight='bold')

plt.show()
```
